```python
import yfinance as yf

# Define the ticker symbols for the companies you want to analyze
ticker_symbols = [
    # Banking and Finance
    'HDFCBANK.NS',
    'ICICIBANK.NS',
    'AXISBANK.NS',
    
    # Automobile
    'MARUTI.NS',
    'BAJAJ-AUTO.NS',
    'HEROMOTOCO.NS',
    
    # Information Technology
    'TCS.NS',
    'WIPRO.NS',
    'TECHM.NS',
    
    # Energy
    'ONGC.NS',
    'BPCL.NS',
    
    # Pharmaceuticals
    'SUNPHARMA.NS',
    'DRREDDY.NS',
    'CIPLA.NS',
    
    # Consumer Goods
    'HINDUNILVR.NS',
    'ITC.NS',
    'NESTLEIND.NS',
    
    # Telecommunications
    'BHARTIARTL.NS',
    'IDEA.NS',
    
    # Utilities
    'NTPC.NS',
    'POWERGRID.NS',
    'TATAPOWER.NS',
    
    # Cement
    'ULTRACEMCO.NS',
    'AMBUJACEM.NS',
    'SHREECEM.NS',
    
    # Metals and Mining
    'TATASTEEL.NS',
    'HINDALCO.NS',
    'VEDL.NS'
]

# Fetch historical data
stock_data = yf.download(ticker_symbols, start='2020-01-01', end='2023-1-31')

```

```python
# Save the stock data to a CSV file
stock_data.to_csv('stock_data.csv')
```

```python
import pandas as pd

# Create a DataFrame with the ticker symbols as columns
sector_mapping = pd.DataFrame(columns=ticker_symbols)

# Create a dictionary to map tickers to sectors
sector_mapping_dict = {
    'TATAMOTORS.NS': 'Automobile',
    'RELIANCE.NS': 'Energy',
    'INFY.NS': 'Information Technology',
    'HDFCBANK.NS': 'Banking and Finance',
    'ICICIBANK.NS': 'Banking and Finance',
    'AXISBANK.NS': 'Banking and Finance',
    'MARUTI.NS': 'Automobile',
    'BAJAJ-AUTO.NS': 'Automobile',
    'HEROMOTOCO.NS': 'Automobile',
    'TCS.NS': 'Information Technology',
    'WIPRO.NS': 'Information Technology',
    'TECHM.NS': 'Information Technology',
    'ONGC.NS': 'Energy',
    'BPCL.NS': 'Energy',
    'SUNPHARMA.NS': 'Pharmaceuticals',
    'DRREDDY.NS': 'Pharmaceuticals',
    'CIPLA.NS': 'Pharmaceuticals',
    'HINDUNILVR.NS': 'Consumer Goods',
    'ITC.NS': 'Consumer Goods',
    'NESTLEIND.NS': 'Consumer Goods',
    'BHARTIARTL.NS': 'Telecommunications',
    'IDEA.NS': 'Telecommunications',
    'NTPC.NS': 'Utilities',
    'POWERGRID.NS': 'Utilities',
    'TATAPOWER.NS': 'Utilities',
    'ULTRACEMCO.NS': 'Cement',
    'AMBUJACEM.NS': 'Cement',
    'SHREECEM.NS': 'Cement',
    'TATASTEEL.NS': 'Metals and Mining',
    'HINDALCO.NS': 'Metals and Mining',
    'VEDL.NS': 'Metals and Mining'
}


# Add a 'Sector' column to the sector_mapping DataFrame
sector_mapping['Sector'] = sector_mapping.index.map(sector_mapping_dict)

# Rename the columns in the stock_data DataFrame to avoid conflicts
stock_data.columns = [f'{col}_Stock' for col in stock_data.columns]

# Merge the sector_mapping DataFrame with the stock_data DataFrame
stock_data = stock_data.join(sector_mapping)

```
